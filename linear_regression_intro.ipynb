```python
%matplotlib inline

import pandas as pd
import numpy as np
from datetime import datetime

movies = pd.read_csv('2013_movies.csv', parse_dates={'Date':['ReleaseDate']})
movies['Date'] = movies['Date'].astype(datetime)

movies.head()
```

```python
movies.describe()
```

```python
from matplotlib import pyplot as plt
movies['DomesticTotalGross'].hist()
plt.title('Histogram of Movies by Box Office Revenues')
plt.xlabel('Number of Movies')
plt.show()
```

```python
movies['Budget'].hist()

plt.title('Histogram of Movies by Budget')
plt.xlabel('Budget (hundred million dollars)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# Are budget and domestic total gross related?
plt.scatter(movies.Budget, movies.DomesticTotalGross)

plt.title('How does budget relate to final box office revenue?')
plt.xlabel('Budget')
plt.ylabel('Box office')
```

```python
movies.head()
```

```python
# Get all the movies that have no (known) budget
movies_without_budget = movies[np.isnan(movies.Budget)]
movies_without_budget
```

```python
movies_with_budget = movies.dropna()
movies_with_budget
```

```python
# statsmodels: looks a lot like R
# scikit.learn: doesn't give as much information, but all of other ML stuff is also in scikit.learn

movies_with_budget['Ones'] = 1.0
movies_with_budget.tail()
```

```python
# statsmodels and pandas work nicely together

import statsmodels.formula.api as sm

# We have the ones because statsmodels wants to multiply with this one the constant parameter
# the designers wanted the interface to remain consistent => hence the need for Ones
X = movies_with_budget[['Budget','Ones']]
Y = movies_with_budget['DomesticTotalGross']

# OLS = ordinary least squares
linmodel = sm.OLS(Y,X).fit()
linmodel.summary()
```

Prediction: we're going to make 40 million + 78% of our budget

```python
plt.scatter(movies.Budget, movies.DomesticTotalGross)
plt.title('How does budget relate to final box office recenue?')
plt.xlabel('Budget')
plt.ylabel('Box Office')

plt.plot(movies_with_budget.Budget, linmodel.predict(X), 'r-')
plt.show()
```

```python
# Same thing, but now with scikit.learn
from sklearn.linear_model import LinearRegression

X = movies_with_budget[['Budget']] # X has to be a matrix
Y = movies_with_budget['DomesticTotalGross']

sk_linmodel = LinearRegression()
sk_linmodel.fit(X,Y) # NOTE: X,Y, not Y,X as in statsmodel. Why? Because!

print 'coefficient:\t %g' % sk_linmodel.coef_ # "Internal" 
print 'intercept:  \t %g' % sk_linmodel.intercept_ # "Internal"
print 'R squared:  \t %g' % sk_linmodel.score(X,Y)
```

```python
plt.scatter(X,Y)
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(X, sk_linmodel.predict(X), 'r-')
plt.show()
```

```python
# Now we compare the actual vs. the predicted values
plt.plot(Y, sk_linmodel.predict(X), 'go')
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Model')

plt.plot(Y,Y, 'm--', label="Ideal Prediction") # Magenda dashed line
plt.show()
```

```python
# Now let's go for another regression analysis
movies_with_budget['TitleLength'] = [len(title) for title in movies_with_budget.Title]
movies_with_budget.head()
```

```python
X = movies_with_budget[['Budget','Runtime','TitleLength','Ones']]
Y = movies_with_budget['DomesticTotalGross']

linmodel = sm.OLS(Y,X).fit()
linmodel.summary()
```

```python
# Multi-dimensional regression, but only plot against budget
plt.scatter(movies.Budget, movies.DomesticTotalGross)
plt.title('How does budget relate to final box office recenue?')
plt.xlabel('Budget')
plt.ylabel('Box Office')

plt.plot(movies_with_budget.Budget, linmodel.predict(X), 'rs')
plt.show()
```
