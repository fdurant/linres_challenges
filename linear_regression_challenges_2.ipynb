## Introduction

Our client is MovieMarketingMoneyMaker Incorporated (M4 Inc.), a company specializing in shared-risk marketing investments in the movie industry. M4 Inc. CEO Dino Brangelino is not happy with the returns he made on his 2014 investments in movies he handpicked so carefully. Even though he considers himself an expert in the movie business, he has therefore turned to our .data consulting firm to better understand the factors that influence movie profitability.

Since M4 Inc. is rather risk averse, Mr. Brangelino prefers to invest only after having seen the outcome of the first weekend gross revenues. If possible, he would also like to understand the risk of adopting a pre-release strategy that does not involve this first weekend information.

The request from our client is a mathematical model that predicts the profitability ratio for each new movie.

In this study, movie profitability crucially depends on three values:
- production cost, which is known at release date
- marketing cost, which is unknown at release date, but estimated at 50% of production cost (source: http://entertainment.howstuffworks.com/movie-cost1.htm)
- domestic total gross revenue, which is only known after the fact, of course

Profitability ratio
= domestic total gross revenue / (production cost x marketing cost)
= domestic total gross revenue / (production cost x 1.5)


## Data Preparation

The full historic data set scraped from BoxOfficeMojo.com on Tuesday January 27, 2015 contains 15244 entries. As the target variable (profitability) is fundamentally based on production budget, any decent model must mandatorily include this input variable.

```python
import pandas as pd
import numpy as np
from datetime import datetime, date
moviesDF=pd.read_csv('BoxOfficeMojoMovies.csv', sep=',', header=0, index_col=0, parse_dates=True)
moviesDF['ReleaseDate'] = pd.to_datetime(moviesDF['ReleaseDate'])
print "Original number of entries: %d" % len(moviesDF)
#moviesDF.dtypes
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

moviesDF['yearOfRelease'] = [date.year for date in moviesDF['ReleaseDate']]
moviesDF['withProductionBudget'] = [not pd.isnull(pb) for pb in moviesDF['ProductionBudget']]
moviesDF['withoutProductionBudget'] = [pd.isnull(pb) for pb in moviesDF['ProductionBudget']]

groupByYear = moviesDF.groupby('yearOfRelease')
yearlyWithProductionBudget = groupByYear['withProductionBudget'].sum()
yearlyWithoutProductionBudget = groupByYear['withoutProductionBudget'].sum()

plt.figure(1)
plt.title("# of movies with and without known production budget")
yearlyWithProductionBudget.plot(label='Known production budget')
yearlyWithoutProductionBudget.plot(label='Unknown production budget')
plt.legend(loc=2)
plt.show()
```

Overall, production budget has been badly recorded across time. In absolute terms, however, the peak of _known_ production budgets still lies in the recent 20 years. As a result, we still have enough data to build a profitability prediction model for movies made in the near future. Let's take a closer look at the known production budgets since 1995:

```python
plt.figure(15, figsize=(9,6))
plt.title("# of recent movies with known production budget")
yearlyWithProductionBudget[1995:2015].plot(kind='bar')
plt.ylabel("# movies")
plt.show()
```

At first sight, it looks like we will have a decent amount of movie entries to train a model from. However, to do so, we first want to make sure that no relevant column contains missing values. We also remove outliers for ProductionBudget and DomesticTotalGross, and convert the categorical fields into binary presence/absence ones.

```python
irrelevantColumns = ['Actors','Directors','Writers','Producers','Composers','ForeignTotalGross','withProductionBudget','withoutProductionBudget']
relevantColumns = [x for x in moviesDF.columns]
for c in irrelevantColumns:
    if c in relevantColumns:
        relevantColumns.remove(c)
print "relevantColumns:", relevantColumns
moviesDF2 = moviesDF.copy()
moviesDF2 = moviesDF2.dropna(subset=relevantColumns)
moviesDF2.drop(irrelevantColumns, inplace=True, axis=1)

print "Remaining number of entries after removal of NaN values in relevant columns: %d" % len(moviesDF2)

# Removal of outliers: keep only the ones that are within +3 to -3 standard deviations
# in the columns 'ProductionBudget' and 'DomesticTotalGross'.
moviesDF3 = moviesDF2.copy()
moviesDF3 = moviesDF3[np.abs(moviesDF3.ProductionBudget-moviesDF3.ProductionBudget.mean())<=(3*moviesDF3.ProductionBudget.std())] 
print "Remaining number of entries after removal of ProductionBudget outliers: %d" % len(moviesDF3)

moviesDF4 = moviesDF3.copy()
moviesDF4 = moviesDF4[np.abs(moviesDF4.DomesticTotalGross-moviesDF4.DomesticTotalGross.mean())<=(3*moviesDF4.DomesticTotalGross.std())] 
print "Remaining number of entries after removal of DomesticTotalGross outliers: %d" % len(moviesDF4)

# Convert categorical columns
moviesDF5 = pd.concat([moviesDF4, pd.get_dummies(moviesDF4['MPAARating'])], axis=1)
moviesDF6 = pd.concat([moviesDF5, pd.get_dummies(moviesDF5['ReleaseSeason'])], axis=1)

finalDF = moviesDF6[moviesDF6['yearOfRelease'] >= 1995]
print "Remaining number of entries after removal of NaN values in relevant columns: %d" % len(finalDF)
finalDF.head()
```

```python
totalNrEntries = float(len(finalDF))
print "totalNrEntries = %d" % totalNrEntries
trainingDF = finalDF[finalDF['yearOfRelease'] <= 2011]
nrTrainingEntries = len(trainingDF)
print "nrTrainingEntries = %d" % nrTrainingEntries
testDF = finalDF[finalDF['yearOfRelease'] >= 2012]
nrTestEntries = len(testDF)
print "nrTestEntries = %d" % nrTestEntries
print "Number of training entries: %d (%1.1f%%)" % (nrTrainingEntries, nrTrainingEntries/totalNrEntries*100)
print "Number of test entries: %d (%1.1f%%)" % (nrTestEntries, nrTestEntries/totalNrEntries*100)

```

# Challenge 2

Generate (fake) data that is linearly related to log(x). Basically write an underlying model just like in challenge 1, but instead of a fitted model, you are making this model up. It is of the form B0 + B1log(x) + epsilon. You are making up the parameters. Simulate some data from this model. Then fit two models to it: a) quadratic [second degree polynomial] b) logarithmic [log(x)] (the second one should fit really well, since it has the same form as the underlying model!)

```python
import numpy as np
import math

def makeFakeLogXData(x):
    b0 = 7.5
    b1 = 35
    epsilon = np.random.normal(50, 7, None)
#    print >> sys.stderr, "epsilon = ", epsilon
    result = b0 + b1*math.log(x) + epsilon
    return result

fakeLogXDF = pd.DataFrame({'X' : pd.Series(abs(np.random.randn(5000)))})
fakeLogXDF['Y'] = [makeFakeLogXData(x) for x in fakeLogXDF['X']]

fig = plt.figure(1, figsize=(9,6))

plt.scatter(fakeLogXDF['X'],fakeLogXDF['Y'])
plt.xlabel("abs(np.random.randn)")
plt.ylabel("some linear transformation of log(X)")
plt.show()
```

```python
import statsmodels.formula.api as sm
import math

fakeLogXDF['XSquared'] = [x*x for x in fakeLogXDF['X']]
fakeLogXDF['LogX'] = [math.log(x) for x in fakeLogXDF['X']]
fakeLogXDF['Ones'] = 1

Xsq = fakeLogXDF[['X','XSquared','Ones']]
Ynew = fakeLogXDF['Y']

linmodelSq = sm.OLS(Ynew,Xsq).fit()
linmodelSq.summary()
```

```python
Xlog = fakeLogXDF[['LogX','Ones']]
linmodelLog = sm.OLS(Ynew,Xlog).fit()
linmodelLog.summary()
```

# Challenge 4

For one continuous feature (like budget, choose one that strongly affects the outcome), try polynomial fits from 0th (just constant) to 7th order (highest term x^7). Over the x axis of model degree (8 points), plot: a) training error b) test error c) R squared d) AIC

```python
trainingDF4 = trainingDF
testDF4 = testDF

modelDegrees = range(8) # 0 through 7
modelsByDegree = []

# Build the polynomial regression models of increasing order
for degree in modelDegrees:
    fieldName = 'Xpow%d' % degree

    # Add missing polynomial terms to both training and test dataframe
    trainingDF4[fieldName] = [pow(x,degree) for x in trainingDF4['ProductionBudget']]
    testDF4[fieldName] = [pow(x,degree) for x in testDF4['ProductionBudget']]
    
    # Build and store the actual models
    xFieldNameList = ['Xpow%d' % deg for deg in modelDegrees[0:degree+1]]
    X = trainingDF4[xFieldNameList]
    Y = trainingDF4['DomesticTotalGross']
#    print >> sys.stderr, "Building polynomial model of degree %d ..." % degree,
    model = sm.OLS(Y,X).fit()
#    print >> sys.stderr, "done"
    modelsByDegree.append(model)
    
# Evaluate the models on test data, storing the error measures
errorMeasures = {'rmseTraining':[],
                 'rmseTest':[],
                 'rsquared':[],
                 'aic':[]}
for degree in modelDegrees:
    xFieldNameList = ['Xpow%d' % deg for deg in modelDegrees[0:degree+1]]
    model = modelsByDegree[degree]

    trainingX = trainingDF4[xFieldNameList]
    observedTrainingY = trainingDF4['DomesticTotalGross']
    predictedTrainingY = model.predict(trainingX)
    
    testX = testDF4[xFieldNameList]
    observedTestY = testDF4['DomesticTotalGross']
    predictedTestY = model.predict(testX)

    errorMeasures['rmseTraining'].append(math.sqrt(mean_squared_error(observedTrainingY,predictedTrainingY)))
    errorMeasures['rmseTest'].append(math.sqrt(mean_squared_error(observedTestY,predictedTestY)))
    errorMeasures['rsquared'].append(model.rsquared)
    errorMeasures['aic'].append(model.aic)
    
# Print and plot the error measures
print "root-mean-squared error on training data:", errorMeasures['rmseTraining'], "\n"
print "root-mean-squared error on test data:", errorMeasures['rmseTest'], "\n"
print "RSquared:", errorMeasures['rsquared'], "\n"
print "Akaike information criterion", errorMeasures['aic'], "\n"

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

ax1.plot(modelDegrees, errorMeasures['rmseTraining'])
#ax1.set_title("RMSE on training data")
ax1.set_xlabel('polynomial model degree')
ax1.set_ylabel('RMSE - training')

ax2.plot(modelDegrees, errorMeasures['rmseTest'])
#ax2.set_title("RMSE on test data")
ax2.set_xlabel('polynomial model degree')
ax2.set_ylabel('RMSE - test')
               
ax3.plot(modelDegrees, errorMeasures['rsquared'])
#ax3.set_title("RSquared")
ax3.set_xlabel('polynomial model degree')
ax3.set_ylabel('RSquared')

ax4.plot(modelDegrees, errorMeasures['aic'])
#ax4.set_title("AIC")
ax4.set_xlabel('polynomial model degree')
ax4.set_ylabel('AIC')

plt.tight_layout()
plt.show()
```

Conclusion: for the given data set, the polynomial regression model with degree 1 gives the best result, all other things being equal.

# Challenge 5

Fit a model to only the first 5 of your data points (m=5). Then to first 10 (m=10). Then to first 15 (m=15). In this manner, keep fitting until you fit your entire training set. For each step, calculate the training error and the test error. Plot both (in the same plot) over m. This is called a learning curve.

```python
trainingDF5 = trainingDF
testDF5 = testDF

mValues = []
trainingRMSEs = []
testRMSEs = []

learningStep = 5

m = learningStep
while m <= len(trainingDF5):
    mValues.append(m)
    
    # Build model
    X = trainingDF4[['ProductionBudget','Ones']][0:m]
    Y = trainingDF4['DomesticTotalGross'][0:m]
    model = sm.OLS(Y,X).fit()
        
    trainingX = X
    observedTrainingY = trainingDF4['DomesticTotalGross'][0:m]
    predictedTrainingY = model.predict(trainingX)
    
    testX = testDF4[['ProductionBudget','Ones']][0:m]
    observedTestY = testDF4['DomesticTotalGross'][0:m]
    predictedTestY = model.predict(testX)

    # Evaluate model
    trainingRMSEs.append(math.sqrt(mean_squared_error(observedTrainingY,predictedTrainingY)))
    testRMSEs.append(math.sqrt(mean_squared_error(observedTestY,predictedTestY)))
    
    m = m + learningStep
    
# Plot learning curve
print mValues[0:5]
print trainingRMSEs[0:5]
print testRMSEs[0:5]

fig = plt.figure(1, figsize=(9,6))
plt.title('Learning curves for training and test set')
plt.plot(mValues, trainingRMSEs, label='on training set')
plt.plot(mValues, testRMSEs, label='on test set')
plt.xlabel('number of model training instances')
plt.ylabel('root-mean-squared error')
plt.legend()
plt.show()
```

With m approximately < 100, the models are grossly overfit to the training data, and therefore basically useless.

With m approximately between 100 and 450, the models score better on the test than on the training data.

With m approximately > 450, the models score worse on the test than on the training data. This is to be expected, since the test data are (by design) unknown to the model.
