## Introduction

Our client is MovieMarketingMoneyMaker Incorporated (M4 Inc.), a company specializing in shared-risk marketing investments in the movie industry. M4 Inc. CEO Dino Brangelino is not happy with the returns he made on his 2014 investments in movies he handpicked so carefully. Even though he considers himself an expert in the movie business, he has therefore turned to our .data consulting firm to better understand the factors that influence movie profitability.

Since M4 Inc. is rather risk averse, Mr. Brangelino prefers to invest only after having seen the outcome of the first weekend gross revenues. If possible, he would also like to understand the risk of adopting a pre-release strategy that does not involve this first weekend information.

The request from our client is a mathematical model that predicts the profitability ratio for each new movie.

In this study, movie profitability crucially depends on three values:
- production cost, which is known at release date
- marketing cost, which is unknown at release date, but estimated at 50% of production cost (source: http://entertainment.howstuffworks.com/movie-cost1.htm)
- domestic total gross revenue, which is only known after the fact, of course

Profitability ratio
= domestic total gross revenue / (production cost * marketing cost)
= domestic total gross revenue / (production cost * 1.5)


## Data Preparation

The full historic data set scraped from BoxOfficeMojo.com on Tuesday January 27, 2015 contains 15244 entries. As the target variable (profitability) is fundamentally based on production budget, any decent model must mandatorily include this input variable.

```python
import pandas as pd
import numpy as np
from datetime import datetime, date
moviesDF=pd.read_csv('BoxOfficeMojoMovies.csv', sep=',', header=0, index_col=0, parse_dates=True)
moviesDF['ReleaseDate'] = pd.to_datetime(moviesDF['ReleaseDate'])
print "Original number of entries: %d" % len(moviesDF)
#moviesDF.dtypes
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

moviesDF['yearOfRelease'] = [date.year for date in moviesDF['ReleaseDate']]
moviesDF['withProductionBudget'] = [not pd.isnull(pb) for pb in moviesDF['ProductionBudget']]
moviesDF['withoutProductionBudget'] = [pd.isnull(pb) for pb in moviesDF['ProductionBudget']]

groupByYear = moviesDF.groupby('yearOfRelease')
yearlyWithProductionBudget = groupByYear['withProductionBudget'].sum()
yearlyWithoutProductionBudget = groupByYear['withoutProductionBudget'].sum()

plt.figure(1)
plt.title("# of movies with and without known production budget")
yearlyWithProductionBudget.plot(label='Known production budget')
yearlyWithoutProductionBudget.plot(label='Unknown production budget')
plt.legend(loc=2)
plt.show()
```

Overall, production budget has been badly recorded across time. In absolute terms, however, the peak of _known_ production budgets still lies in the recent 20 years. As a result, we still have enough data to build a profitability prediction model for movies made in the near future. Let's take a closer look at the known production budgets since 1995:

```python
plt.figure(1)
plt.title("# of recent movies with known production budget")
yearlyWithProductionBudget[1995:2015].plot(kind='bar')
plt.ylabel("# movies")
plt.show()
```

At first sight, it looks like we will have a decent amount of movie entries to train a model from. However, to do so, we first want to make sure that no relevant column contains missing values. We also remove outliers for ProductionBudget and DomesticTotalGross, and convert the categorical fields into binary presence/absence ones.

```python
irrelevantColumns = ['Actors','Directors','Writers','Producers','Composers','ForeignTotalGross','withProductionBudget','withoutProductionBudget']
relevantColumns = [x for x in moviesDF.columns]
for c in irrelevantColumns:
    if c in relevantColumns:
        relevantColumns.remove(c)
print "relevantColumns:", relevantColumns
moviesDF2 = moviesDF.dropna(subset=relevantColumns)
moviesDF2.drop(irrelevantColumns, inplace=True, axis=1)

print "Remaining number of entries after removal of NaN values in relevant columns: %d" % len(moviesDF2)

# Removal of outliers: keep only the ones that are within +3 to -3 standard deviations
# in the columns 'ProductionBudget' and 'DomesticTotalGross'.
moviesDF3 = moviesDF2[np.abs(moviesDF2.ProductionBudget-moviesDF2.ProductionBudget.mean())<=(3*moviesDF2.ProductionBudget.std())] 
print "Remaining number of entries after removal of ProductionBudget outliers: %d" % len(moviesDF3)

moviesDF4 = moviesDF3[np.abs(moviesDF3.DomesticTotalGross-moviesDF3.DomesticTotalGross.mean())<=(3*moviesDF3.DomesticTotalGross.std())] 
print "Remaining number of entries after removal of DomesticTotalGross outliers: %d" % len(moviesDF4)

# Convert categorical columns
moviesDF5 = pd.concat([moviesDF4, pd.get_dummies(moviesDF4['MPAARating'])], axis=1)
moviesDF6 = pd.concat([moviesDF5, pd.get_dummies(moviesDF5['ReleaseSeason'])], axis=1)

finalDF = moviesDF6[moviesDF6['yearOfRelease'] >= 1995]
print "Remaining number of entries after removal of NaN values in relevant columns: %d" % len(finalDF)
finalDF.head()
```

```python
totalNrEntries = float(len(finalDF))
print "totalNrEntries = %d" % totalNrEntries
trainingDF = finalDF[finalDF['yearOfRelease'] <= 2011]
nrTrainingEntries = len(trainingDF)
print "nrTrainingEntries = %d" % nrTrainingEntries
testDF = finalDF[finalDF['yearOfRelease'] >= 2012]
nrTestEntries = len(testDF)
print "nrTestEntries = %d" % nrTestEntries
print "Number of training entries: %d (%1.1f%%)" % (nrTrainingEntries, nrTrainingEntries/totalNrEntries*100)
print "Number of test entries: %d (%1.1f%%)" % (nrTestEntries, nrTestEntries/totalNrEntries*100)

```

# Challenge 1

Fit a model to your data. Now you know the coeeficients (the beta values).

Write a python function to simulate outcomes. 

(The sigma of the normal noise distribution will be the std deviation of the residuals.) For the same observed input variables, simulate the outcome. Plot the observed data and the simulated data.

```python
import statsmodels.formula.api as sm

trainingDF['Ones'] = 1.0

X = trainingDF[['ProductionBudget','OWAvgGrossPerTheater','Ones']]
Y = trainingDF['DomesticTotalGross']

linmodel = sm.OLS(Y,X).fit()
linmodel.summary()
```

Prediction: roughly speaking, we're going to make 81% of our production budget + (464 * the OWAvgGrossPerTheater) + 9.6 million

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

testDF['Ones'] = 1.0
X2 = testDF[['ProductionBudget','OWAvgGrossPerTheater','Ones']]

fig = plt.figure(15, figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')

surf = ax.scatter(testDF['ProductionBudget'], testDF['OWAvgGrossPerTheater'], testDF['DomesticTotalGross'])
plt.title('Production budget & opening weekend average gross versus total domestic gross')

ax.set_xlim(0.0, max(testDF['ProductionBudget']))
ax.set_ylim(0.0, max(testDF['OWAvgGrossPerTheater']))
ax.set_zlim(0.0, max(testDF['DomesticTotalGross']))

ax.set_xlabel("ProductionBudget")
ax.set_ylabel("OWAvgGrossPerTheater")
ax.set_zlabel("DomesticTotalGross")

ax.plot(testDF.ProductionBudget, testDF.OWAvgGrossPerTheater, linmodel.predict(X2), 'rs')
plt.show()
```

We now compare the observed outcome from the test set with their corresponding predicted values:

```python
observedDomesticTotalGross = testDF['DomesticTotalGross']
testInputVariables = testDF[['ProductionBudget','OWAvgGrossPerTheater','Ones']]
predictedDomesticTotalGross = linmodel.predict(testInputVariables)

fig = plt.figure(15, figsize=(9,6))
plt.plot(observedDomesticTotalGross, predictedDomesticTotalGross, 'go')
plt.title('Actual vs. predicted DomesticTotalGross (TEST set)')
plt.xlabel('Observed')
plt.ylabel('Predicted')

plt.plot(observedDomesticTotalGross,observedDomesticTotalGross, 'm--', label="Ideal Prediction") # Magenta dashed line
plt.show()
```

Finally, we also plot the observed versus predicted profitability ratio.

```python
def calculateProfitabilityRatio(domesticTotalGross,productionBudget,marketingMultiplier=1.5):
    '''returns the profitability ratio'''
    return domesticTotalGross / (productionBudget * marketingMultiplier)

testDF['observedProfitabilityRatio'] = testDF.apply(lambda row: calculateProfitabilityRatio(row['DomesticTotalGross'],row['ProductionBudget']), axis=1)
obsProfRatio = testDF.loc[:,['ProductionBudget','OWAvgGrossPerTheater','DomesticTotalGross','observedProfitabilityRatio']].sort(columns='observedProfitabilityRatio', ascending=False)
obsProfRatio.head(5)
```

```python
import sys
def predictProfitabilityRatio(linearModel, *args):
    productionBudget = args[0]
    predictedDomesticTotalGross = linearModel.predict(args)
    result = calculateProfitabilityRatio(predictedDomesticTotalGross, productionBudget)
#    result = 1
    return result

testDF['predictedProfitabilityRatio'] = testDF.apply(lambda row: predictProfitabilityRatio(linmodel,row['ProductionBudget'],row['OWAvgGrossPerTheater'],row['Ones']), axis=1)
predProfRatio = testDF.loc[:,['ProductionBudget','OWAvgGrossPerTheater','DomesticTotalGross','observedProfitabilityRatio','predictedProfitabilityRatio']].sort(columns='predictedProfitabilityRatio', ascending=False)
predProfRatio.head(5)
```

Here's what we can learn so far from the current model:

* its most extreme overpredictions in absolute terms (ratio 10 and more) occur with movies that actually have observed profitability ratios close to zero. It would be most interesting to identify other features that can single out these flopped movies.

* movies with observed profitability ratios of 2 and more are underpredicted, but in a somewhat _systematic_ way. This is suggested by the (manually produced) green line, which is a simple linear transformation from the magenta one. It is trivial to capture this partial regularity by adapting the original model in the first place.
