## Introduction

Our client is MovieMarketingMoney Incorporated (MMM Inc.), a company specializing in shared-risk marketing investments with and in the movie industry. MMM Inc. CEO Dino Brangelino was not happy with the returns he made on his 2014 investments in the movies he handpicked so carefully. Therefore, even though he considers himself an expert in the movie business, he has turned to our .data consulting firm to better understand the factors that influence movie profitability.

Since MMM Inc. is rather risk averse, Mr. Brangelino prefers to invest only after having seen the outcome of the first weekend gross revenues. If possible, he would also like to understand the risk of adopting a pre-release strategy that does not involve this first weekend gross revenue information.

The request from our client is a mathematical model that predicts the profitability ratio for each new movie.

In this study, movie profitability crucially depends on three values:
- production cost, which is known at release date
- marketing cost, which is unknown at release date, but estimated at 50% of production cost (source: http://entertainment.howstuffworks.com/movie-cost1.htm)
- domestic total gross revenue, which is only known after the fact, of course

Profitability ratio
= domestic total gross revenue / (production cost * marketing cost)
= domestic total gross revenue / (production cost * 1.5)


## Data Preparation

The full historic data set scraped from BoxOfficeMojo.com on Tuesday January 27, 2015 contains 15244 entries. As the target variable (profitability) is fundamentally based on production budget, any decent model must mandatorily include this input variable.

```python
import pandas as pd
import numpy as np
from datetime import datetime, date
moviesDF=pd.read_csv('BoxOfficeMojoMovies.csv', sep=',', header=0, index_col=0, parse_dates=True)
moviesDF['ReleaseDate'] = pd.to_datetime(moviesDF['ReleaseDate'])
print "Original number of entries: %d" % len(moviesDF)
#moviesDF.dtypes
```

```python
%matplotlib inline
from matplotlib import pyplot as plt

moviesDF['yearOfRelease'] = [date.year for date in moviesDF['ReleaseDate']]
moviesDF['withProductionBudget'] = [not pd.isnull(pb) for pb in moviesDF['ProductionBudget']]
moviesDF['withoutProductionBudget'] = [pd.isnull(pb) for pb in moviesDF['ProductionBudget']]

groupByYear = moviesDF.groupby('yearOfRelease')
yearlyWithProductionBudget = groupByYear['withProductionBudget'].sum()
yearlyWithoutProductionBudget = groupByYear['withoutProductionBudget'].sum()

plt.figure(1)
plt.title("# of movies with and without known production budget")
yearlyWithProductionBudget.plot(label='Known production budget')
yearlyWithoutProductionBudget.plot(label='Unknown production budget')
plt.legend(loc=2)
plt.show()
```

Overall, production budget has been badly recorded across time. In absolute terms, however, the peak of _known_ production budgets still lies in the recent 20 years. As a result, we still have enough data to build a profitability prediction model for movies made in the near future. Let's take a closer look at the known production budgets since 1995:

```python
plt.figure(1)
plt.title("# of recent movies with known production budget")
yearlyWithProductionBudget[1995:2015].plot(kind='bar')
plt.ylabel("# movies")
plt.show()
```

At first sight, it looks like we will have a decent amount of movie entries to train a model from. However, to do so, we first want to make sure that no relevant column contains missing values. We also convert the categorical field into binary presence/absence ones.

```python
irrelevantColumns = ['Actors','Directors','Writers','Producers','Composers','ForeignTotalGross','withProductionBudget','withoutProductionBudget']
relevantColumns = [x for x in moviesDF.columns]
for c in irrelevantColumns:
    relevantColumns.remove(c)
print "relevantColumns:", relevantColumns
moviesDF = moviesDF.dropna(subset=relevantColumns)
moviesDF.drop(irrelevantColumns, inplace=True, axis=1)

# Convert categorical columns
moviesDF = pd.concat([moviesDF, pd.get_dummies(moviesDF['MPAARating'])], axis=1)
moviesDF = pd.concat([moviesDF, pd.get_dummies(moviesDF['ReleaseSeason'])], axis=1)

finalDF = moviesDF[moviesDF['yearOfRelease'] >= 1995]
print "Remaining number of entries after removal of NaN values in relevant columns: %d" % len(finalDF)
finalDF.head()
```

```python
totalNrEntries = float(len(finalDF))
print "totalNrEntries = %d" % totalNrEntries
trainingDF = finalDF[finalDF['yearOfRelease'] <= 2011]
nrTrainingEntries = len(trainingDF)
print "nrTrainingEntries = %d" % nrTrainingEntries
testDF = finalDF[finalDF['yearOfRelease'] >= 2012]
nrTestEntries = len(testDF)
print "nrTestEntries = %d" % nrTestEntries
print "Number of training entries: %d (%1.1f%%)" % (nrTrainingEntries, nrTrainingEntries/totalNrEntries*100)
print "Number of test entries: %d (%1.1f%%)" % (nrTestEntries, nrTestEntries/totalNrEntries*100)

```

# Challenge 1

Fit a model to your data. Now you know the coeeficients (the beta values).

Write a python function to simulate outcomes. 

(The sigma of the normal noise distribution will be the std deviation of the residuals.) For the same observed input variables, simulate the outcome. Plot the observed data and the simulated data.

```python
import statsmodels.formula.api as sm

trainingDF['Ones'] = 1.0

X = trainingDF[['ProductionBudget','OWAvgGrossPerTheater','Ones']]
Y = trainingDF['DomesticTotalGross']

linmodel = sm.OLS(Y,X).fit()
linmodel.summary()
```

Prediction: roughly speaking, we're going to make 102% of our production budget + (636 * the OWAvgGrossPerTheater) + 3.5 million

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

testDF['Ones'] = 1.0
X2 = testDF[['ProductionBudget','OWAvgGrossPerTheater','Ones']]

fig = plt.figure(15, figsize=(15,10))
ax = fig.add_subplot(111, projection='3d')

surf = ax.scatter(testDF['ProductionBudget'], testDF['OWAvgGrossPerTheater'], testDF['DomesticTotalGross'])
plt.title('Production budget & opening weekend average gross versus total domestic gross')

ax.set_xlim(0.0, max(testDF['ProductionBudget']))
ax.set_ylim(0.0, max(testDF['OWAvgGrossPerTheater']))
ax.set_zlim(0.0, max(testDF['DomesticTotalGross']))

ax.set_xlabel("ProductionBudget")
ax.set_ylabel("OWAvgGrossPerTheater")
ax.set_zlabel("DomesticTotalGross")

ax.plot(testDF.ProductionBudget, testDF.OWAvgGrossPerTheater, linmodel.predict(X2), 'rs')
plt.show()
```

We now compare the observed outcome from the test set with their corresponding predicted values:

```python
observedDomesticTotalGross = testDF['DomesticTotalGross']
testInputVariables = testDF[['ProductionBudget','OWAvgGrossPerTheater','Ones']]
predictedDomesticTotalGross = linmodel.predict(testInputVariables)

fig = plt.figure(15, figsize=(15,10))
plt.plot(observedDomesticTotalGross, predictedDomesticTotalGross, 'go')
plt.title('Actual vs. predicted DomesticTotalGross (TEST set)')
plt.xlabel('Observed')
plt.ylabel('Predicted')

plt.plot(observedDomesticTotalGross,observedDomesticTotalGross, 'm--', label="Ideal Prediction") # Magenta dashed line
plt.show()
```
